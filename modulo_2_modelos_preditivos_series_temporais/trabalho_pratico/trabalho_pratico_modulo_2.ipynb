```python
# importando as bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://pycourse.s3.amazonaws.com/bike-sharing.csv", parse_dates=['datetime'])
df
```

```python
# info
df.info()
```

```python
# Qual a média da coluna windspeed? 
df['windspeed'].mean()
```

```python
# Qual a média da coluna temp? 
df['temp'].mean()
```

```python
# Quantos registros de locações existem para o 
# ano de 2011 (número de linhas, número de colunas)? 
df[df['datetime'].dt.year == 2011]
```

```python
# Quantos registros de locações existem para o 
# ano de 2012 (número de linhas, número de colunas)? 
df[df['datetime'].dt.year == 2012]
```

```python
df_2011 = df[df['datetime'].dt.year == 2011]
df_2011['total_count'].sum()
```

```python
# Quantas locações de bicicletas foram efetuadas em 2012? 
df_2012 = df[df['datetime'].dt.year == 2012]
df_2012['total_count'].sum()
```

```python
# Qual estação do ano contém a maior média de locações de bicicletas? 
# Qual estação do ano contém a menor média de locações de bicicletas? 
df2 = df.groupby('season').mean()
print(df2['total_count'])
```

```python
# Qual horário do dia contém a maior média de locações de bicicletas? 
df3 = df.groupby('hour').mean()
print(df3['total_count'])
```

```python
print(df3[df3['total_count'] == df3['total_count'].max()]['total_count'])
```

```python
# Qual horário do dia contém a menor média de locações de bicicletas?
print(df3[df3['total_count'] == df3['total_count'].min()]['total_count'])
```

```python
# Qual dia da semana contém a maior média de locações de bicicletas? 
df4 = df.groupby('weekday').mean()
print(df4['total_count'])
```

```python
print(df4[df4['total_count'] == df4['total_count'].max()]['total_count'])
```

```python
# Que dia da semana contém a menor média de locações de bicicletas?
print(df4[df4['total_count'] == df4['total_count'].min()]['total_count'])
```

```python
# Às quartas-feiras (weekday = 3), qual horário do dia
# contém a maior média de locações de bicicletas?
df5 = df[df['weekday'] == 3].groupby('hour').mean()
print(df5['total_count'])
```

```python
print(df5[df5['total_count'] == df5['total_count'].max()]['total_count'])
```

```python
# Aos sábados (weekday = 6), qual horário do dia 
# contém a maior média de locações de bicicletas? 
df6 = df[df['weekday'] == 6].groupby('hour').mean()
print(df6['total_count'])
```

```python
print(df6[df6['total_count'] == df6['total_count'].max()]['total_count'])
```
